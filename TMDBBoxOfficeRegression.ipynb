## Load packages

```python
import numpy as np
import pandas as pd
import ast
from collections import Counter
from scipy.stats import boxcox

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
from wordcloud import WordCloud
```

## Data Loading and overviewing of categorical variables

### Load data

```python
train_path = './Input/train.csv'
train = pd.read_csv(train_path, index_col='id')

# for the kaggle kernel
# train_path = '../input/tmdb-box-office-prediction/train.csv'
# train = pd.read_csv(train_path, index_col='id')

dict_columns = ['belongs_to_collection', 'genres', 'production_companies',
                'production_countries', 'spoken_languages', 'Keywords', 'cast', 'crew']

# function taken from the kernel: https://www.kaggle.com/gravix/gradient-in-a-box
def text_to_dict(df):
    for column in dict_columns:
        df[column] = df[column].apply(lambda x: {} if pd.isna(x) else ast.literal_eval(x))
    return df

train = text_to_dict(train)

train_view = train.copy()
```

```python
# using 'boxplot method' to find anomalies
def get_anomalies(data):

    anomalies_indices = []
    
    for i in data:
        data_series = data.loc[:, i]

        q1 = np.quantile(data_series, 0.25)
        q3 = np.quantile(data_series, 0.75)
        iqr = q3 - q1
        upper_limit = q1 + 1.5 * iqr
        lower_limit = q1 - 1.5 * iqr

        anomalies_indices = anomalies_indices + data_series[(data_series > upper_limit) | (data_series < lower_limit)].index.tolist()

    # remove duplicates
    anomalies_indices = list(set(anomalies_indices))
    return anomalies_indices
```

## Feature Engineering

I usually load the test dataset after selecting the final algorithm, but in this notebook I will need it beforehand to fill in zero values for budget. But first I have to create the same variables that I created earlier.

```python
def correct_release_date(x):
    if int(x[-2:]) >= 0 and int(x[-2:]) <= 19: 
        d = x[:-2] + '20' + x[-2:]
    else:
        d = x[:-2] + '19' + x[-2:]
    return d 
```

```python
def get_new_variables(df):
    # belongs_to_collection
    df['collection_name'] = df.belongs_to_collection.apply(lambda x: x[0]['name'] if x != {} else 0)
    df['in_collection'] = df.belongs_to_collection.apply(lambda x: 1 if x != {} else 0)

    # genres
    list_of_genres = list(train_view.genres.apply(lambda x: [i['name'] for i in x] if x != {} else []))
    df['num_genres'] = df.genres.apply(lambda x: len(x) if x != {} else 0)
    df['all_genres'] = df.genres.apply(lambda x: [i['name'] for i in x] if x != {} else [])
    top_genres = [i[0] for i in Counter([i for j in list_of_genres for i in j]).most_common(15)]
    for g in top_genres:
        df['genres_' + str(g)] = df.all_genres.apply(lambda x: 1 if g in x else 0)

    # production_companies
    list_of_production_companies = list(train_view.production_companies.apply(lambda x: [i['name'] for i in x] if x != 0 else []))
    df['num_production_companies'] = df.production_companies.apply(lambda x: len(x) if x != {} else 0)
    df['all_production_companies'] = df.production_companies.apply(lambda x: [i['name'] for i in x] if x != {} else [])
    top_production_companies = [x[0] for x in Counter([i for j in list_of_production_companies for i in j]).most_common(30)]
    for c in top_production_companies:
        df['production_companies_' + str(c)] = df.all_production_companies.apply(lambda x: 1 if c in x else 0)

    # production_countries
    list_of_production_countries = list(train_view.production_countries.apply(lambda x: [i['name'] for i in x] if x != 0 else []))
    df['num_production_countries'] = df.production_countries.apply(lambda x: len(x) if x != {} else 0)
    df['all_production_countries'] = df.production_countries.apply(lambda x: [i['name'] for i in x] if x != {} else [])
    top_production_countries = [x[0] for x in Counter([i for j in list_of_production_countries for i in j]).most_common(25)]

    for c in top_production_countries:
        df['production_countries_' + str(c)] = df.all_production_countries.apply(lambda x: 1 if c in x else 0)

    # spoken_languages
    list_of_spoken_languages = list(train_view.spoken_languages.apply(lambda x: [i['name'] for i in x] if x != {} else []))
    df['num_spoken_languages'] = df.spoken_languages.apply(lambda x: len(x) if x != {} else 0)
    df['all_spoken_languages'] = df.spoken_languages.apply(lambda x: [i['name'] for i in x] if x != {} else [])
    top_spoken_languages = [x[0] for x in Counter([i for j in list_of_spoken_languages for i in j]).most_common(20)]

    for c in top_spoken_languages:
        df['spoken_languages_' + str(c)] = df.all_spoken_languages.apply(lambda x: 1 if c in x else 0)

    # keywords
    list_of_keywords = list(train_view.Keywords.apply(lambda x: [i['name'] for i in x] if x != {} else []))
    df['num_keywords'] = df.Keywords.apply(lambda x: len(x) if x != {} else 0)
    df['all_keywords'] = df.Keywords.apply(lambda x: [i['name'] for i in x] if x != {} else [])
    top_Keywords = [x[0] for x in Counter([i for j in list_of_keywords for i in j]).most_common(20)]

    for c in top_Keywords:
        df['Keywords_' + str(c)] = df.all_keywords.apply(lambda x: 1 if c in x else 0)

    # cast
    list_of_cast_names = list(train_view.cast.apply(lambda x: [i['name'] for i in x] if x != {} else []))
    list_of_cast_characters = list(train_view.cast.apply(lambda x: [i['character'] for i in x] if x != {} else []))
    df['num_cast'] = df.cast.apply(lambda x: len(x) if x != {} else 0)
    for g in list('012'):
        df['cast_gender_' + str(g)] = df.cast.apply(lambda x: sum([1 for i in x if i['gender'] == int(g)]))

    top_cast_names = [x[0] for x in Counter([i for j in list_of_cast_names for i in j]).most_common(20)]
    for c in top_cast_names:
        df['cast_name_' + str(c)] = df.cast.apply(lambda x: 1 if c in str(x) else 0)

    top_cast_characters = [x[0] for x in Counter([i for j in list_of_cast_characters for i in j if i != '']).most_common(20)]
    for c in top_cast_characters:
        df['cast_character_' + str(c)] = df.cast.apply(lambda x: 1 if c in str(x) else 0)

    # crew
    list_of_crew_names = list(train_view.crew.apply(lambda x: [i['name'] for i in x] if x != {} else []))
    list_of_crew_jobs = list(train_view.crew.apply(lambda x: [i['job'] for i in x] if x != {} else []))
    df['num_crew'] = df.crew.apply(lambda x: len(x) if x != {} else 0)
    for g in list('012'):
        df['crew_gender_' + str(g)] = df.crew.apply(lambda x: sum([1 for i in x if i['gender'] == int(g)]))

    top_crew_names = [x[0] for x in Counter([i for j in list_of_crew_names for i in j]).most_common(15)]
    for c in top_crew_names:
        df['crew_name_' + str(c)] = df.crew.apply(lambda x: 1 if c in str(x) else 0)

    top_crew_jobs = [x[0] for x in Counter([i for j in list_of_crew_jobs for i in j]).most_common(15)]
    for c in top_crew_jobs:
        df['crew_job_' + str(c)] = df.crew.apply(lambda x: 1 if c in str(x) else 0)

    # release_date
    df['corrected_release_date'] = pd.to_datetime(df.release_date.apply(correct_release_date))

    for t in ['year', 'month', 'day', 'weekday', 'quarter']:
         df['release_date_' + t] = getattr(df['corrected_release_date'].dt, t)  
        
    # homepage
    df['has_homepage'] = df.homepage.apply(lambda x: 1 if pd.notnull(x) else 0)

    
    return df
```

## Preliminary assessment of machine learning algorithms

```python
from scipy.special import inv_boxcox

def inv_y(transformed_y):
    boxcox_y = boxcox_num_values['revenue'][0].reshape(-1, 1)
    y = combined_df.revenue[combined_df.revenue.notnull()]

    scaler.fit(boxcox_y)
    inversed_y = scaler.inverse_transform(transformed_y)

    original_y = inv_boxcox(inversed_y, boxcox_lmbdas['revenue'])
#     original_y = inv_boxcox(transformed_y, boxcox_lmbdas['revenue'])
#     when revenue is close to zero, the inverse of the Box-Cox transformation returns missing value
    if np.sum(np.isnan(original_y)) != 0:
        original_y[np.isnan(original_y)] = 0
    return original_y
```

```python
from sklearn.metrics import mean_absolute_error

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor

mae_compare = pd.Series()

# Lasso
lasso_model = Lasso(alpha=0.0005, random_state=0)
lasso_model.fit(X_train.values, y_train.values)
lasso_val_preds = lasso_model.predict(X_val)
lasso_val_mae = mean_absolute_error(inv_y(lasso_val_preds), inv_y(y_val))
mae_compare['Lasso'] = lasso_val_mae

# Ridge
ridge_model = Ridge(random_state=0)
ridge_model.fit(X_train, y_train)
ridge_val_preds = ridge_model.predict(X_val)
ridge_val_mae = mean_absolute_error(inv_y(ridge_val_preds), inv_y(y_val))
mae_compare['Ridge']= ridge_val_mae

# ElasticNet
elastic_model = ElasticNet(alpha=0.02, l1_ratio=0.7, random_state=0)
elastic_model.fit(X_train, y_train)
elastic_val_preds = elastic_model.predict(X_val)
elastic_val_mae = mean_absolute_error(inv_y(elastic_val_preds), inv_y(y_val))
mae_compare['ElasticNet'] = elastic_val_mae

# KNeighborsRegressor
knn_model = KNeighborsRegressor()
knn_model.fit(X_train, y_train)
knn_val_preds = knn_model.predict(X_val)
knn_val_mae = mean_absolute_error(inv_y(knn_val_preds), inv_y(y_val))
mae_compare['KNeighborsRegressor']= knn_val_mae

# Random Forest
rf_model = RandomForestRegressor(random_state=0)
rf_model.fit(X_train, y_train)
rf_val_preds = rf_model.predict(X_val)
rf_val_mae = mean_absolute_error(inv_y(rf_val_preds), inv_y(y_val))
mae_compare['RandomForestRegressor']= rf_val_mae

# GradientBoostingRegressor
gb_model = GradientBoostingRegressor(n_estimators=300, learning_rate=0.05, 
                                      max_depth=4, random_state=0)
gb_model.fit(X_train, y_train)
gb_val_preds = gb_model.predict(X_val)
gb_val_mae = mean_absolute_error(inv_y(gb_val_preds), inv_y(y_val))
mae_compare['GradientBoostingRegressor'] = gb_val_mae

# AdaBoostRegressor
ab_model = AdaBoostRegressor(n_estimators=300, learning_rate=0.05,
                        random_state=0)
ab_model.fit(X_train, y_train)
ab_val_preds = ab_model.predict(X_val)
ab_val_mae = mean_absolute_error(inv_y(ab_val_preds), inv_y(y_val))
mae_compare['AdaBoostRegressor'] = ab_val_mae

# XGBRegressor 
xgb = XGBRegressor(n_estimators = 1000, 
                    learning_rate=0.05, random_state=0)
xgb.fit(X_train.values, y_train.values, 
       early_stopping_rounds=5,
       eval_set=[(X_val.values, y_val.values)],
       verbose=False)
xgb_val_pred = xgb.predict(X_val.values)
xgb_val_mae = mean_absolute_error(inv_y(xgb_val_pred), inv_y(y_val))
mae_compare['XGBRegressor'] = xgb_val_mae

# display all mae values
print('Algorithms with the lowest mean absolute error:')
mae_compare.sort_values()
```
